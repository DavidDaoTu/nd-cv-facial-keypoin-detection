## Keypoint detection

Now, we'll loop over each detected face in an image (again!) only this time, you'll transform those faces in Tensors that your CNN can accept as input images.

### TODO: Transform each detected face into an input Tensor

You'll need to perform the following steps for each detected face:
1. Convert the face from RGB to grayscale
2. Normalize the grayscale image so that its color range falls in [0,1] instead of [0,255]
3. Rescale the detected face to be the expected square size for your CNN (224x224, suggested)
4. Reshape the numpy image into a torch image.

### TODO: Detect and display the predicted keypoints

After each face has been appropriately converted into an input Tensor for your network to see as input, you'll wrap that Tensor in a Variable() and can apply your `net` to each face. The ouput should be the predicted the facial keypoints. These keypoints will need to be "un-normalized" for display, and you may find it helpful to write a helper function like `show_keypoints`. You should end up with an image like the following with facial keypoints that closely match the facial features on each individual face:

```python
# from data_load import Rescale, RandomCrop, Normalize, ToTensor
class MyTransform:
    
    @staticmethod
    def ReScale(img, output_size):
        """Rescale an image to desired output_size.
        Args:
            + img: source image in numpy type
            + output_size: int or tuple (new_height, new_width)
        """
        img_h = img.shape[0]
        img_w = img.shape[1]
        
        if isinstance(output_size, int):
            ### Retain the edges ratio between old height/ old width
            if img_h > img_w:
                new_h, new_w = output_size * img_h/ img_w, output_size
            else:
                #img_h < img_w
                new_h, new_w = output_size, output_size * img_w/ img_h
        else:
            # tuple (new_height, new_width)
            new_h, new_w = output_size
        
        ### Rounded
        new_h, new_w = int(new_h), int(new_w)
        
        ### Resize
        return cv2.resize(img, (new_w, new_h))
    
    @staticmethod
    def ToFloatTensor(img):
        """Convert the ndarray img to a float tensor (pytorch)."""
        # if image has no grayscale color channel, add one
        if (len(img.shape) == 2):
            # add third color dim
            img = img.reshape(img.shape[0], img.shape[1], 1)
        
        # Swap color axis because
        # numpy image: H x W x C - (0, 1, 2)
        # torch image: C x H x W - (2, 0, 1)
        img = img.transpose((2, 0, 1))
        
        return torch.from_numpy(img).type(torch.FloatTensor)
    
    @staticmethod
    def ToCNN(one_img):
        """Converts a torch tensor image to conv layer of the CNN that 
        contain batch_size at the first position"""
        return torch.unsqueeze(one_img, dim=0)
 
def display_output_keypoints(faces, kpts):
    fig = plt.figure(figsize=(20, 10))
    for i, face in enumerate(faces):
        ax = fig.add_subplot(1, len(faces), i+1)
        # undo normalization for keypoint image
        keypoint_img = kpts[i] * 50.0 + 100
        
        plt.imshow(face, cmap='gray') # plot detected face
        plt.scatter(keypoint_img[:, 0], keypoint_img[:, 1], s=56, marker='.', c='m')
        plt.axis('off')
    
    plt.show()
    
    
    
####    
image_copy = np.copy(image)
detected_faces = [] #numpy image
predicted_kpts = [] #numpy image

# loop over the detected faces from your haar cascade
for (x,y,w,h) in faces:
    
    # Select the region of interest that is the face in the image 
    roi = image_copy[y:y+h, x:x+w]
    
    ## TODO: Convert the face region from RGB to grayscale
    roi = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)

    ## TODO: Normalize the grayscale image so that its color range falls in [0,1] instead of [0,255]
    roi = roi / 255.0
    
    ## TODO: Rescale the detected face to be the expected square size for your CNN (224x224, suggested)
    roi = MyTransform.ReScale(roi, 224)
    detected_faces.append(roi) # adding a detected image to our list
    
    ## TODO: Reshape the numpy image shape (H x W x C) into a torch image shape (C x H x W)
    roi = MyTransform.ToFloatTensor(roi)
        
    ## TODO: Make facial keypoint predictions using your loaded, trained network 
    ## perform a forward pass to get the predicted facial keypoints
    # Convert image to Convolution matching
    roi = MyTransform.ToCNN(roi)
    
    # forward pass to get net output facial keypoints (output_pts)
    output_pts = net.forward(roi)
    
    # reshape to batch_size x 68 x 2 pts
    output_pts = output_pts.view(output_pts.size()[0], 68, -1)
    
    # Obtain the output key points image & convert to numpy()
    output_pts = output_pts[0].data.numpy()
    print(output_pts.shape)
    predicted_kpts.append(output_pts)
    
    

## TODO: Display each detected face and the corresponding keypoints        
display_output_keypoints(detected_faces, predicted_kpts)
```
